```python
# exercise 4
plt.figure(figsize=(15, 8))
cfd = nltk.ConditionalFreqDist(
    (target, fileid) for fileid in state_union.fileids()
    for w in state_union.words(fileid)
    for target in ["men", "women", "people"] if w.lower() == target)
cfd.plot(title="Usage of the terms 'men', 'women', and 'people'")
plt.grid(True)
plt.show()
```

```python
# exercise 11
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd_modals = nltk.ConditionalFreqDist(
    (genre, word.lower())
    for genre in genres
    for word in brown.words(categories=genre)
    if word.lower() in modals
)
print("Modal Verb Distribution Across Genres:")
cfd_modals.tabulate(conditions=genres, samples=modals)

pronouns = ['i', 'you', 'he', 'she', 'it', 'we', 'they']
cfd_pronouns = nltk.ConditionalFreqDist(
    (genre, word.lower())
    for genre in genres
    for word in brown.words(categories=genre)
    if word.lower() in pronouns
)
print("\nPersonal Pronoun Distribution Across Genres:")
cfd_pronouns.tabulate(conditions=genres, samples=pronouns)
```

The modeal verb distribution clearly represent the style of the text. For example in text about hobbies the word "can" is dominat as these are probabily informal and pretty direct pieces of text. Other big diffenreces are in the news and romance genres where respectively the words "will" and "could" are the most used out of all the modals.

Regarding personal pronoun the situation is similar, as we can distinguish clearly genres with higher usage of certain words. For example in romance the personal pronouns "I" and "he" are mostly used indicating that most of these text are narrated in first person by a woman character, or again in the hobbies category the most used personal pronoun is "you" indicating a more direct and informal text.

```python
# exercise 16
print(f"{'Genre':<16} | {'Tokens':<10} | {'Types':<10} | {'Lexical Diversity':<10}")
print("-" * 65)

for genre in brown.categories():
    tokens = len(brown.words(categories=genre))
    types = len(set(w.lower() for w in brown.words(categories=genre)))
    lexical_diversity = tokens / types if types > 0 else 0
    print(f"{genre:<16} | {tokens:<10} | {types:<10} | {lexical_diversity:<18.2f}")
```

Humor genre has the lowest diverity meaning that most of the words used in these types of text are often repeating and from a short pool of words

```python
# exercise 23 
def plot_zipf(text_words, title="Zipf's Law Plot"):
    """
    Calculates and plots word frequency against word rank on a log-log scale.
    """
    fdist = nltk.FreqDist(w.lower() for w in text_words if w.isalpha())
    frequencies = [count for _, count in fdist.most_common()]
    ranks = range(1, len(frequencies) + 1)
    
    # Plot on a log-log scale
    plt.figure(figsize=(12, 7))
    plt.plot(ranks, frequencies)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('Rank (log scale)')
    plt.ylabel('Frequency (log scale)')
    plt.title(title)
    plt.grid(True)
    plt.show()

plot_zipf(text1, title="Zipf's Law for text1")

num_chars = len(''.join(text1))
alphabet = 'abcdefghijklmnopqrstuvwxyz '
random_text = ''.join(random.choice(alphabet) for _ in range(num_chars))
random_words = random_text.split()
plot_zipf(random_words, title="Zipf's Law for Random Text")
```

At the high frequency the line is stright representing the most common workds, at the low-frequency end, where the rank approaches N, the plot takes a staircase shape, because each word only occurs a few times or just once.

The same plot for random words does not follwo a straight line. It is more a S curve shape indicationg that the frequency of each word is much more evenly distrubued.

Zipf's law is a power law distribution like the scale free power law distrubution of the Barbasi-Albert netwroks. It works in a similar way in network theory as a node, hub, with many links is very rare respect to the most common nodes which have just a few links. Similarly a frequency of a word is inversely proprotional to its rank, creating fer very common words and many rare words. 

The internet itselfs, more specifically the distribution of website traffic, is another example of power law. Very fre websites (Google, Wikipedia etc.) have the majority of visits while most other have way less traffic. Another example is the network of links between web pages (in links). They are all examples of the "rich-gets-richer" phenomenon that is at the core of BA networks.
